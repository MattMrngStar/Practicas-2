```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def numerico_euler (t,x):
  """
  esta función evalúa dx/dy para Euler 
  Parámetros: t, x(variable independiente, variable dependiente) 
  retorna: La función evaluada en los valores de t & x
  """
  dif = (1/math.log(2))*(1/t)    #Esta es la ecuacion diferencial que se va a evaluar 
  #c/k es igual 1 / ln(2)
  return dif
```

```python
def numerico_euler2 (t2,x2):
  """
  esta función evalúa dx/dt para Euler 
  Parámetros: t, x(variable independiente, variable dependiente) 
  retorna: La función evaluada en los valores de t & x
  """
  dif2 = (1/math.log(2))*(1/t2)    #Esta es la ecuacion diferencial que se va a evaluar
  return dif2
```

```python
def analitica(t,x):
  """
  Esta función evalúa la solución analítica (a mano)
  Parámetros: x,y (variable independiente, variable dependiente)
  retorna: La función evaluada en los valores de t & x
  """
  #sol = (1/math.log((1+math.sqrt(5))/2))*(math.log((1/((((1+math.sqrt(5))/2)**2)-1))+t))+1   #Esta es la solución analítica de la ED
  #sol  = (2.078*math.log(0.618+t))+1
  sol = (math.log(t) / math.log(2)) + 2
  return sol



```
